OPENCV를 사용한다.

cv2.imshow 함수를 cv2_imshow로 변경한다.

```python
import cv2
from google.colab.patches import cv2_imshow
```

이미지를 cv2.imread함수로 불러오고

cv2_imshow함수로 이미지 출력

```python
img ='/content/drive/MyDrive/pm.jpg'
image = cv2.imread(img)
cv2_imshow(image)
```

이미지에서 테두리만 따오기위해 cv2.GaussianBlur함수 사용.

```python
blur = cv2.GaussianBlur(image, ksize=(3,3), sigmaX=0)
ret, thresh1 = cv2.threshold(blur, 127, 255, cv2.THRESH_BINARY)
edged = cv2.Canny(blur, 10, 250)
cv2_imshow(edged)
```
